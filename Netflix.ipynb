```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sqlite3
```

```python
data = pd.read_csv('netflix.csv')
```

```python
data.isnull().sum()
data.fillna(0,inplace=True)
```

```python
data['duration'] = data['duration'].str.replace(' min','')
data['duration'] = data['duration'].str.replace(' Seasons','')
data['duration'] = data['duration'].str.replace(' Season','')
data['duration']
```

```python
data['type'].value_counts().plot(kind='bar', color=['#DC143C', '#B22222'])
plt.title('TV Shows vs Movies')
plt.show()
```

```python
data.groupby('release_year').size().plot(kind='line', color='#B22222')
plt.title('Year-wise Content Release')
plt.show()
```

```python
top_countries = data['country'].value_counts().head(10)
top_countries.plot(kind='bar', color=['#DC143C', '#B22222', '#DC143C', '#B22222', '#DC143C', '#B22222', '#DC143C', '#B22222', '#DC143C', '#B22222'])
plt.title('Top 10 Countries with Most Content')
plt.show()
```

```python
data['listed_in'] = data['listed_in'].str.split(',')
exploded_genres = data.explode('listed_in')
top_genres = exploded_genres['listed_in'].value_counts().head(10)
top_genres.plot(kind='bar', color=['#DC143C', '#B22222', '#DC143C', '#B22222', '#DC143C', '#B22222', '#DC143C', '#B22222', '#DC143C', '#B22222'])
plt.title('Top Genres')
plt.show()
```

```python
data['rating'].value_counts().plot(kind='bar', color=['#DC143C', '#B22222', '#DC143C', '#B22222', '#DC143C', '#B22222', '#DC143C', '#B22222', '#DC143C', '#B22222'])
plt.title('Distribution of Content Ratings')
plt.show()
```
